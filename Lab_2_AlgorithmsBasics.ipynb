```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
start=time.time()
print(func5(16))
end=time.time()
print("running time:",end-start)
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
start=time.time()
print(func8([3,1,2,7,8,6]))
end=time.time()
print("running time:",end-start)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
    total = 0
    for element in arr:
        total += element
    return total
```

```python

arr = [12, 3, 4, 15,45,78,450]

start=time.time()
print(sum_arr(arr))
end=time.time()
print("running time:",end-start)
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,x):
  for i in range(len(arr)):
    if x==arr[i]:
      return i
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr,x):
  start=0
  end=len(arr)-1
  while start<=end:
    mid=(end-start)//2
    if(arr[mid]==x):
      return mid
    elif mid<x:
      end=mid-1
    else:
      start=mid+1
```
