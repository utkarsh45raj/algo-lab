# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    m = [[0 for _ in range(n)] for _ in range(n)]
    s = [[0 for _ in range(n)] for _ in range(n)]

    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (
                    m[i][k] +
                    m[k+1][j] +
                    p[i-1] * p[k] * p[j]
                )

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k

    return m, s
```

```python
def print_optimal_parenthesization(s, i, j):
    if i == j:
        return f"A{i}"
    else:
        left = print_optimal_parenthesization(s, i, s[i][j])
        right = print_optimal_parenthesization(s, s[i][j] + 1, j)
        return f"({left}{right})"

def calculate_naive_cost(p):
    n = len(p)
    if n <= 2:
        return 0

    cost = 0

    cost += p[0] * p[1] * p[2]

    current_cost = 0
    if n > 2:
        current_cost += p[0] * p[1] * p[2] # A1x(A2x(A3...)) (A1A2)A3

    for i in range(3, n):
        current_cost += p[0] * p[i-1] * p[i]

    return current_cost
```

```python
p_exercise = [5, 10, 3, 12, 5, 50, 6]
n_exercise = len(p_exercise)

m_exercise, s_exercise = matrix_chain_multiplication(p_exercise)

min_cost = m_exercise[1][n_exercise-1]
print(f"Minimum multiplication cost: {min_cost}")

optimal_parenthesization = print_optimal_parenthesization(s_exercise, 1, n_exercise-1)
print(f"Optimal parenthesization: {optimal_parenthesization}")

naive_cost = calculate_naive_cost(p_exercise)
print(f"Naive multiplication cost (left-to-right): {naive_cost}")

multiplications_saved = naive_cost - min_cost
print(f"Multiplications saved: {multiplications_saved}")

print("\nDP Table (m):")
for row in m_exercise[1:]:
    print(row[1:])

print("\nSplit Table (s):")
for row in s_exercise[1:]:
    print(row[1:])
```
